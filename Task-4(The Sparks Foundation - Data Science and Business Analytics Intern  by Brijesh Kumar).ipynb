### Visualizing the data

### 1. Years wise attack


1. No of attacks in each year

```python
attacks=df['Year'].value_counts(dropna=True).sort_index().to_frame().reset_index().rename(columns={'index':'Years','Year':'attacks'}).set_index('Years')
attacks.head(10)
```

```python
attacks.plot(kind='bar',figsize=(16,7),color='blue',fontsize=15)
plt.title('Timeline of attacks',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of attacks',fontsize=15)
```

 - most number of attacks (16903)in 2014
 - least number of attacks (471) in 1971

### Region wise Attacks
  1. Distribution of terrorist attacks over the years in regions

```python
reg=pd.crosstab(df.Year,df.Region)
reg.head(10)
```

```python
reg.plot(kind='area',figsize=(20,10),stacked=False,alpha=0.5)
plt.title('Regin wise attack',fontsize=15)
plt.xlabel('Regins',fontsize=15)
plt.ylabel('Number of attack',fontsize=15)
```

### total terrorist attack in each region from 1970-2017

```python
regt=reg.transpose()
regt['Total']=regt.sum(axis=1)
ra=regt['Total'].sort_values(ascending=False)
ra
```

```python
ra.plot(kind='bar',figsize=(20,10),fontsize=10)
plt.title('Total number of attack in each region 1970-2017',fontsize=15)
plt.xlabel('Regions',fontsize=15)
plt.ylabel('Number of attack',fontsize=15)
```
